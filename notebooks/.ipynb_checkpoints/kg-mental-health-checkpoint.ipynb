```python
def format_age(born_date):
    age = 2022 - born_date

    if age < 18:
        return "0-17"
    elif age < 25:
        return "18-24"
    elif age < 35:
        return "25-34"
    elif age < 45:
        return "35-44"
    elif age < 55:
        return "45-54"
    else:
        return "55 or more"
```
