```python
import networkx as nx
```

```python
graph = nx.Graph()
```

```python
nodes = open("../data/graph-nodes-ex01.txt","r")

for node in nodes:
    graph.add_node(node[0])

edges = open("../data/graph-edges-ex01.txt","r")

for edge in edges:
    node1, node2 = edge.split(",")
    graph.add_edge(node1[0], node2[0])
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw(graph, with_labels=True, font_weight='bold')
```

# Cálculo das métricas

Centralidade Local

```python
localy = nx.degree_centrality(graph)
localy
```

```python
import matplotlib.cm as cm
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt
```

```python
cmap = cm.Reds 
norm = mcolors.Normalize(vmin=0, vmax=1)
colors = [cmap(norm(localy[node])) for node in graph.nodes]

plt.figure(figsize=(10, 8))
pos = nx.spring_layout(graph)
nx.draw(
    graph, pos, with_labels=True, node_color=colors, cmap=cmap, 
    edge_color='gray', node_size=800, font_size=10
)

sm = cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm, ax=plt.gca(), label="Centrality")

plt.title("Graph Colored by Centrality")
plt.show()
```
